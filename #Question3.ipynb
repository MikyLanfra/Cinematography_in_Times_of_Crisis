Cleaning: removing punctuation and converting to lowercase

```python
df = pd.read_csv('data/movies_with_summaries.csv')
df.head()
```

```python
# Function to clean text data
def clean_text(text):
    # Remove special characters and digits
    cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Convert text to lowercase
    text = cleaned_text.lower()
    return text
```

```python
df['cleaned_summary'] = df['summary'].apply(clean_text)
```

```python
minidf=df.head(1000)
```

```python
synopses=minidf["cleaned_summary"].tolist()

##df['tokenized_summaries'] = df['cleaned_summary'].apply(word_tokenize)
```

```python
# Process the documents: tokenize and remove stopwords
processed_docs = []
for synopsis in synopses:
    tokens = word_tokenize(synopsis.lower())
    # Keep alphanumeric tokens that are not stopwords
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    processed_docs.append(filtered_tokens)
```

```python
# Initialize the Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()

# Perform Sentiment Analysis
sentiment_scores = []
for tokens in processed_docs:
    # Join tokens back into a string
    text = ' '.join(tokens)
    # Get sentiment scores for the text
    score = analyzer.polarity_scores(text)
    sentiment_scores.append(score)
```

```python
# Convert the list of sentiment scores to a DataFrame
sentiment_df = pd.DataFrame(sentiment_scores)

# Reset the index of 'minidf' to align with 'sentiment_df'
minidf = minidf.reset_index(drop=True)

# Concatenate the sentiment scores with the original DataFrame
minidf = pd.concat([minidf, sentiment_df], axis=1)

# Display the relevant columns
print(minidf[['cleaned_summary', 'neg', 'neu', 'pos', 'compound']])
```
